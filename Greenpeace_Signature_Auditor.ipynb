This script automates the verification of petition signatures to ensure high data quality for CRM integration.

```python
# --- GREENPEACE SIGNATURE AUDITOR ---
# Objective: Audit 1,000 petition signatures for data quality

# 1. Our "Dataset" (A mix of valid and invalid emails)
signatures = [
    "petersari@gmail.com", "info@greenpeace.org", "invalid-email.com",
    "kenya_supporter@yahoo.com", "wrong@@top.me", "nairobi.eco@gmail.com",
    "fake_user", "ocean_advocate1@gmail.com", "test@domain", "earth@protect.org"
]

# 2. Tracking Variables
valid_list = []
invalid_list = []
total_checked = 0

print("Starting Audit...")

# 3. The Audit Loop
for email in signatures:
    total_checked += 1

    # Simple Logic: An email MUST have an '@' and a '.' to be potentially valid
    if "@" in email and "." in email:
        valid_list.append(email)
    else:
        invalid_list.append(email)

# 4. The Final Report
print("-" * 30)
print(f"Audit Complete!")
print(f"Total Signatures Checked: {total_checked}")
print(f"Valid Signatures: {len(valid_list)}")
print(f"Invalid Signatures Flagged: {len(invalid_list)}")
print("-" * 30)
print(f"Data Status: READY FOR HUBSPOT UPLOAD")
```

```python
import matplotlib.pyplot as plt

# 1. Data from our previous audit
labels = ['Valid Signatures', 'Invalid Signatures']
sizes = [len(valid_list), len(invalid_list)] # Using the lists from your audit
colors = ['#2ecc71', '#e74c3c'] # Greenpeace Green and Warning Red
explode = (0.1, 0)  # "Explode" the first slice to highlight valid data

# 2. Create the Pie Chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

# 3. Add Title and "Greenpeace Vibe"
plt.title('Greenpeace Petition Data Quality Audit', fontsize=15, fontweight='bold')
plt.axis('equal')  # Ensures the pie is a circle

# 4. Show the magic!
plt.show()
```
